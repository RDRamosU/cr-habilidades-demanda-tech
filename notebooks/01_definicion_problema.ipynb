## 4. Brechas anticipadas

A partir de la revisión preliminar de las fuentes se identifican
tres brechas estructurales:

| Brecha | Descripción | Fuente |
|--------|-------------|--------|
| **Inglés insuficiente** | Todos los perfiles tech exigen B1-C1, nivel que muchos graduados no alcanzan | MTSS 2025 |
| **Frameworks modernos** | Alta demanda de Angular, React, .NET Core, Flutter, AWS, Azure — escasos entre graduados | MTSS 2025 |
| **Certificaciones internacionales** | ITIL, CCNA, Scrum, AWS muy valoradas pero el sistema universitario no las certifica | MTSS 2025 |
| **Habilidades blandas** | Trabajo en equipo interdisciplinario y compromiso con calidad son las más valoradas por empleadores | CONARE 2023 |
