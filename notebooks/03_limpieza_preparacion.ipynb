## 3. Verificación de calidad de datos

```python
print("=== VERIFICACIÓN DE CALIDAD ===\n")

for nombre, df in [('Ocupaciones', df_ocupaciones),
                   ('Habilidades', df_habilidades),
                   ('Competencias', df_competencias),
                   ('Brechas', df_brechas)]:
    nulos = df.isnull().sum().sum()
    duplicados = df.duplicated().sum()
    print(f"{nombre}:")
    print(f"  Filas: {len(df)} | Columnas: {len(df.columns)}")
    print(f"  Nulos: {nulos} | Duplicados: {duplicados}")
    print()

# Verificar consistencia nivel inglés
print("=== CONSISTENCIA NIVEL INGLÉS ===")
print(df_ocupaciones[['ocupacion', 'nivel_ingles',
                       'nivel_ingles_num']].to_string(index=False))

# Verificar categorías únicas
print("\n=== CATEGORÍAS DE HABILIDADES ===")
print(df_habilidades['categoria_habilidad'].value_counts().to_string())

print("\n=== NIVELES REQUERIDOS ===")
print(df_habilidades['nivel_requerido'].value_counts().to_string())
```

## 4. Dataset consolidado — habilidades por categoría

```python
# Habilidades transversales — aparecen en más de una ocupación
df_transversales = (df_habilidades
    .groupby(['habilidad', 'categoria_habilidad'])
    .agg(
        n_ocupaciones=('ocupacion', 'count'),
        ocupaciones=('ocupacion', lambda x: ' / '.join(sorted(x))),
        niveles=('nivel_requerido', lambda x: ' / '.join(sorted(set(x))))
    )
    .reset_index()
    .sort_values('n_ocupaciones', ascending=False)
)

print("=== HABILIDADES TRANSVERSALES (en >1 ocupación) ===\n")
print(df_transversales[df_transversales['n_ocupaciones'] > 1][
    ['habilidad', 'categoria_habilidad',
     'n_ocupaciones', 'niveles']
].to_string(index=False))

# Habilidades por categoría
print("\n=== DISTRIBUCIÓN POR CATEGORÍA ===")
print(df_habilidades.groupby('categoria_habilidad')
      .size().sort_values(ascending=False).to_string())
```

## 5. Dataset integrado — brechas con nivel de impacto ponderado

```python
# Añadir columna de prioridad de atención
mapa_cobertura = {'No': 3, 'Parcialmente': 2, 'Sí': 1}
df_brechas['cobertura_num'] = df_brechas[
    'cubre_sistema_universitario'].map(mapa_cobertura)

# Score de prioridad = impacto * (4 - cobertura)
df_brechas['score_prioridad'] = (
    df_brechas['impacto_num'] * df_brechas['cobertura_num']
)

df_brechas_sorted = df_brechas.sort_values(
    'score_prioridad', ascending=False)

print("=== BRECHAS ORDENADAS POR PRIORIDAD ===\n")
print(df_brechas_sorted[[
    'brecha', 'impacto',
    'cubre_sistema_universitario',
    'score_prioridad'
]].to_string(index=False))

print(f"\nBrecha más urgente: "
      f"{df_brechas_sorted.iloc[0]['brecha']} "
      f"(score: {df_brechas_sorted.iloc[0]['score_prioridad']})")
```

```python
# Exportar todos los datasets procesados
df_ocupaciones.to_csv("ocupaciones_tech_mtss_2025.csv", index=False)
df_habilidades.to_csv("habilidades_tech_mtss_2025.csv", index=False)
df_competencias.to_csv("competencias_empleadores_conare_2023.csv", index=False)
df_brechas.to_csv("brechas_talento_tech_cr.csv", index=False)
df_transversales.to_csv("habilidades_transversales.csv", index=False)

print("=== DATASETS EXPORTADOS ===")
print(f"  ocupaciones_tech_mtss_2025.csv    → {len(df_ocupaciones)} registros ✓")
print(f"  habilidades_tech_mtss_2025.csv    → {len(df_habilidades)} registros ✓")
print(f"  competencias_conare_2023.csv      → {len(df_competencias)} registros ✓")
print(f"  brechas_talento_tech_cr.csv       → {len(df_brechas)} registros ✓")
print(f"  habilidades_transversales.csv     → {len(df_transversales)} registros ✓")
print(f"\nSiguiente paso: Notebook 04 — Visualizaciones y hallazgos")
```
